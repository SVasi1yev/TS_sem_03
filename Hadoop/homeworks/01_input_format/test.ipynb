```python
import zlib
import re
```

```python
def split(string):
    SPLIT_REGEX = re.compile(r'[A-Za-zА-Яа-я0-9]+')
    words = re.findall(SPLIT_REGEX, string)
    return words
```

```python
f_idx = open('data/11e6c777-38e4-4584-86f5-2e50c4144854.pkz.idx', 'rb')
```

```python
f_data = open('data/11e6c777-38e4-4584-86f5-2e50c4144854.pkz', 'rb')
```

```python
python_dict = {}
a = f_idx.read(4)
while len(a) > 0:
    s = set()
    b = int.from_bytes(a, 'little', signed=False)
    c = f_data.read(b)
    d = zlib.decompress(c, zlib.MAX_WBITS)
    words = split(d.decode().lower())
    for word in words:
        if word not in s:
            s.add(word)
            if word not in python_dict:
                python_dict[word] = 1
            else:
                python_dict[word] += 1
    a = f_idx.read(4)
```
