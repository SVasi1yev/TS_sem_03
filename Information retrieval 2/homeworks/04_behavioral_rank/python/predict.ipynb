```python
global_f = 'global_features.tsv'
train_f = 'train_query_features.tsv'
test_f = 'test_query_features.tsv'
marks_f = 'train.marks.tsv'
sample_f = 'sample.csv'
```

```python
import pandas as pd
import numpy as np
```

```python
global_headers = [
      'url_id', 'global_shows', 'global_clicks', 
      'global_pos1shows', 'global_pos2shows',
      'global_pos3shows', 'global_pos4shows',
      'global_pos5shows', 'global_pos1clicks',
      'global_pos2clicks', 'global_pos3clicks',
      'global_pos4clicks', 'global_pos5clicks',
      'global_satisfied', 'global_aver_pos'
    ]

global_features_df = pd.read_csv(global_f, sep='\t', header=None)
global_features_df.columns = global_headers
```

```python
global_features_df.head(5)
```

```python
query_headers = [
        'url_id', 'query_id', 'query_shows', 'query_clicks',
        'query_pos1shows', 'query_pos2shows',
        'query_pos3shows', 'query_pos4shows',
        'query_pos5shows', 'query_pos1clicks',
        'query_pos2clicks', 'query_pos3clicks',
        'query_pos4clicks', 'query_pos5clicks',
        'pos_based_shows', 'cascade_shows', 'cascade_clicks', 
        'dbn_shows', 'dbn_clicks', 'dbn_satisfied',
        'query_aver_pos'
    ]

train_features_df = pd.read_csv(train_f, sep='\t', header=None)
train_features_df.columns = query_headers

test_features_df = pd.read_csv(test_f, sep='\t', header=None)
test_features_df.columns = query_headers
```

```python
train_features_df.head(5)
```

```python
test_features_df.head(5)
```

```python
marks_headers = ['query_id', 'url_id', 'mark']
marks_df = pd.read_csv(marks_f, sep='\t', header=None)
marks_df.columns = marks_headers
```

```python
marks_df.head(5)
```

```python
train_df = marks_df.copy()
train_df = pd.merge(
    train_df, train_features_df, how='left',
    left_on=['query_id', 'url_id'], right_on=['query_id', 'url_id']
)
train_df = pd.merge(
    train_df, global_features_df, how='left',
    left_on=['url_id'], right_on=['url_id']
)

alpha = 0.1

train_df['global_ctr'] = train_df['global_clicks'] / train_df['global_shows']
train_df['global_pos1ctr'] = train_df['global_pos1clicks'] / train_df['global_pos1shows']
train_df['global_pos2ctr'] = train_df['global_pos2clicks'] / train_df['global_pos2shows']
train_df['global_pos3ctr'] = train_df['global_pos3clicks'] / train_df['global_pos3shows']
train_df['global_pos4ctr'] = train_df['global_pos4clicks'] / train_df['global_pos4shows']
train_df['global_pos5ctr'] = train_df['global_pos5clicks'] / train_df['global_pos5shows']
train_df['global_s'] = train_df['global_satisfied'] / train_df['global_clicks']
train_df['query_ctr'] = train_df['query_clicks'] / train_df['query_shows']
train_df['query_pos1ctr'] = train_df['query_pos1clicks'] / train_df['query_pos1shows']
train_df['query_pos2ctr'] = train_df['query_pos2clicks'] / train_df['query_pos2shows']
train_df['query_pos3ctr'] = train_df['query_pos3clicks'] / train_df['query_pos3shows']
train_df['query_pos4ctr'] = train_df['query_pos4clicks'] / train_df['query_pos4shows']
train_df['query_pos5ctr'] = train_df['query_pos5clicks'] / train_df['query_pos5shows']
train_df['pos_based'] = train_df['query_clicks'] / train_df['pos_based_shows']
train_df['cascade'] = train_df['cascade_clicks'] / train_df['cascade_shows']
train_df['dbn_a'] = (train_df['dbn_clicks'] + alpha) / (train_df['dbn_shows'] + 2 * alpha)
train_df['dbn_s'] = (train_df['dbn_satisfied'] + alpha) / (train_df['dbn_clicks'] + 2 * alpha)
train_df['dbn'] = train_df['dbn_a'] * train_df['dbn_s']

train_df = train_df.sort_values(by='query_id')
```

```python
features = list(train_df.columns[3:])
```

```python
filtered_train_df = train_df.dropna(thresh=4)
filtered_train_df
```

```python
sample_headers = ['query_id', 'url_id']
sample_df = pd.read_csv(sample_f, sep=',')
sample_df.columns = sample_headers
```

```python
sample_df.head(5)
```

```python
test_df = sample_df.copy()
test_df = pd.merge(
    test_df, test_features_df, how='left',
    left_on=['query_id', 'url_id'], right_on=['query_id', 'url_id']
)
test_df = pd.merge(
    test_df, global_features_df, how='left',
    left_on=['url_id'], right_on=['url_id']
)

test_df['global_ctr'] = test_df['global_clicks'] / test_df['global_shows']
test_df['global_pos1ctr'] = test_df['global_pos1clicks'] / test_df['global_pos1shows']
test_df['global_pos2ctr'] = test_df['global_pos2clicks'] / test_df['global_pos2shows']
test_df['global_pos3ctr'] = test_df['global_pos3clicks'] / test_df['global_pos3shows']
test_df['global_pos4ctr'] = test_df['global_pos4clicks'] / test_df['global_pos4shows']
test_df['global_pos5ctr'] = test_df['global_pos5clicks'] / test_df['global_pos5shows']
test_df['global_s'] = test_df['global_satisfied'] / test_df['global_clicks']
test_df['query_ctr'] = test_df['query_clicks'] / test_df['query_shows']
test_df['query_pos1ctr'] = test_df['query_pos1clicks'] / test_df['query_pos1shows']
test_df['query_pos2ctr'] = test_df['query_pos2clicks'] / test_df['query_pos2shows']
test_df['query_pos3ctr'] = test_df['query_pos3clicks'] / test_df['query_pos3shows']
test_df['query_pos4ctr'] = test_df['query_pos4clicks'] / test_df['query_pos4shows']
test_df['query_pos5ctr'] = test_df['query_pos5clicks'] / test_df['query_pos5shows']
test_df['pos_based'] = test_df['query_clicks'] / test_df['pos_based_shows']
test_df['cascade'] = test_df['cascade_clicks'] / test_df['cascade_shows']
test_df['dbn_a'] = (test_df['dbn_clicks'] + alpha) / (test_df['dbn_shows'] + 2 * alpha)
test_df['dbn_s'] = (test_df['dbn_satisfied'] + alpha) / (test_df['dbn_clicks'] + 2 * alpha)
test_df['dbn'] = test_df['dbn_a'] * test_df['dbn_s']

test_df = test_df.sort_values(by="query_id")
```

```python
filtered_test_df = test_df.dropna(thresh=3)
filtered_test_df
```

```python
train_queries = set(filtered_train_df.query_id.tolist())
np.random.seed(42)
valid_train_queries = np.random.choice(list(train_queries), int(0.8 * len(train_queries))).tolist()
valid_test_queries = list(train_queries - set(valid_train_queries))
```

```python
valid_train_df = filtered_train_df[filtered_train_df.query_id.apply(lambda x: x in valid_train_queries)]
valid_test_df = filtered_train_df[filtered_train_df.query_id.apply(lambda x: x in valid_test_queries)]
```

```python
from catboost import *
valid_train_pool = Pool(
    data=valid_train_df[features].values,
    label=valid_train_df.mark.values / 5,
    group_id=valid_train_df.query_id.values
)

valid_test_pool = Pool(
    data=valid_test_df[features].values,
    label=valid_test_df.mark.values / 5,
    group_id=valid_test_df.query_id.values
)
full_pool = Pool(
    data=filtered_train_df[features].values,
    label=filtered_train_df.mark.values / 5,
    group_id=filtered_train_df.query_id.values
)
```

```python
model_params = {
    'iterations': 5000,
    'custom_metric': ['NDCG', 'PFound', 'AverageGain:top=5'],
    'loss_function': 'YetiRankPairwise',
    'verbose': True,
    'random_seed': 42,
    'depth':4,
    'subsample':0.9,
    'bootstrap_type':'Bernoulli'
}
```

```python
valid_model = CatBoost(model_params)
valid_model.fit(valid_train_pool, eval_set=[valid_train_pool, valid_test_pool])
```

```python
model = CatBoost(model_params)
model.fit(full_pool, eval_set=[full_pool])
```

```python
test_pool = Pool(data=filtered_test_df[features].values)
filtered_test_df.loc[:, "pred_mark"] = model.predict(test_pool)
result = filtered_test_df.sort_values(by=['query_id',"pred_mark"], ascending = [True, False])
```

```python
with open('sub_cat0.txt', 'w') as f:
    f.write("QueryId,DocumentId\n")
    for query_id in set(result.query_id.tolist()):
        cur_urls = result[result.query_id==query_id].url_id.tolist()
        counter = 0
        used_urls = set()
        for url_id in cur_urls:
            counter += 1
            used_urls.add(url_id)
            f.write("{},{}\n".format(query_id,url_id))
            if counter == 5: 
                break
        if counter < 5:
            available_urls = set(test_df[test_df.query_id==query_id].url_id.tolist()) - used_urls
            if len(available_urls) == 0:
                print("ERROR")
            for url_id in available_urls:
                counter += 1
                f.write("{},{}\n".format(query_id,url_id))
                if counter == 5:
                    break
```
