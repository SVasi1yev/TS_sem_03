```python
def proc_features(df):
    alpha = 0.1

    df['global_url_ctr'] = df['global_url_clicks'] / df['global_url_shows']
    df['global_url_pos1ctr'] = df['global_url_pos1clicks'] / df['global_url_pos1shows']
    df['global_url_pos2ctr'] = df['global_url_pos2clicks'] / df['global_url_pos2shows']
    df['global_url_pos3ctr'] = df['global_url_pos3clicks'] / df['global_url_pos3shows']
    df['global_url_pos4ctr'] = df['global_url_pos4clicks'] / df['global_url_pos4shows']
    df['global_url_pos5ctr'] = df['global_url_pos5clicks'] / df['global_url_pos5shows']
    df['global_url_pos_based'] = df['global_url_clicks'] / df['global_url_pos_based_shows']
    df['global_url_cascade'] = df['global_url_cascade_clicks'] / df['global_url_cascade_shows']
    df['global_url_dbn_a'] = (df['global_url_dbn_clicks'] + alpha) / (df['global_url_dbn_shows'] + 2 * alpha)
    df['global_url_dbn_s'] = (df['global_url_dbn_satisfied'] + alpha) / (df['global_url_dbn_clicks'] + 2 * alpha)
    df['global_url_dbn'] = df['global_url_dbn_a'] * df['global_url_dbn_s']
    
    df['query_url_ctr'] = df['query_url_clicks'] / df['query_url_shows']
    df['query_url_pos1ctr'] = df['query_url_pos1clicks'] / df['query_url_pos1shows']
    df['query_url_pos2ctr'] = df['query_url_pos2clicks'] / df['query_url_pos2shows']
    df['query_url_pos3ctr'] = df['query_url_pos3clicks'] / df['query_url_pos3shows']
    df['query_url_pos4ctr'] = df['query_url_pos4clicks'] / df['query_url_pos4shows']
    df['query_url_pos5ctr'] = df['query_url_pos5clicks'] / df['query_url_pos5shows']
    df['query_url_pos_based'] = df['query_url_clicks'] / df['query_url_pos_based_shows']
    df['query_url_cascade'] = df['query_url_cascade_clicks'] / df['query_url_cascade_shows']
    df['query_url_dbn_a'] = (df['query_url_dbn_clicks'] + alpha) / (df['query_url_dbn_shows'] + 2 * alpha)
    df['query_url_dbn_s'] = (df['query_url_dbn_satisfied'] + alpha) / (df['query_url_dbn_clicks'] + 2 * alpha)
    df['query_url_dbn'] = df['query_url_dbn_a'] * df['query_url_dbn_s']
    
    df['global_host_ctr'] = df['global_host_clicks'] / df['global_host_shows']
    df['global_host_pos1ctr'] = df['global_host_pos1clicks'] / df['global_host_pos1shows']
    df['global_host_pos2ctr'] = df['global_host_pos2clicks'] / df['global_host_pos2shows']
    df['global_host_pos3ctr'] = df['global_host_pos3clicks'] / df['global_host_pos3shows']
    df['global_host_pos4ctr'] = df['global_host_pos4clicks'] / df['global_host_pos4shows']
    df['global_host_pos5ctr'] = df['global_host_pos5clicks'] / df['global_host_pos5shows']
    df['global_host_pos_based'] = df['global_host_clicks'] / df['global_host_pos_based_shows']
    df['global_host_cascade'] = df['global_host_cascade_clicks'] / df['global_host_cascade_shows']
    df['global_host_dbn_a'] = (df['global_host_dbn_clicks'] + alpha) / (df['global_host_dbn_shows'] + 2 * alpha)
    df['global_host_dbn_s'] = (df['global_host_dbn_satisfied'] + alpha) / (df['global_host_dbn_clicks'] + 2 * alpha)
    df['global_host_dbn'] = df['global_host_dbn_a'] * df['global_host_dbn_s']
    
    df['query_host_ctr'] = df['query_host_clicks'] / df['query_host_shows']
    df['query_host_pos1ctr'] = df['query_host_pos1clicks'] / df['query_host_pos1shows']
    df['query_host_pos2ctr'] = df['query_host_pos2clicks'] / df['query_host_pos2shows']
    df['query_host_pos3ctr'] = df['query_host_pos3clicks'] / df['query_host_pos3shows']
    df['query_host_pos4ctr'] = df['query_host_pos4clicks'] / df['query_host_pos4shows']
    df['query_host_pos5ctr'] = df['query_host_pos5clicks'] / df['query_host_pos5shows']
    df['query_host_pos_based'] = df['query_host_clicks'] / df['query_host_pos_based_shows']
    df['query_host_cascade'] = df['query_host_cascade_clicks'] / df['query_host_cascade_shows']
    df['query_host_dbn_a'] = (df['query_host_dbn_clicks'] + alpha) / (df['query_host_dbn_shows'] + 2 * alpha)
    df['query_host_dbn_s'] = (df['query_host_dbn_satisfied'] + alpha) / (df['query_host_dbn_clicks'] + 2 * alpha)
    df['query_host_dbn'] = df['query_host_dbn_a'] * df['query_host_dbn_s']
    
    return df
```
